## Thermal exergy sensitivity

```python
%matplotlib notebook 
# %matplotlib inline
# %matplotlib widget
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Dead state definition:

```python
T0 = 20
```

### The dead state varies between T0-deltaT0 and T0+deltaT0, with deltaT0 one of:

```python
deltaT0s = [-4.0,-3,-2,-1,-0.5]
```

### Graph x array: temperature T

```python
Tlessthan20 = np.arange(-20,20,0.05)
```

```python
# the following was supposed to start from 20.05 to avoid the divide-by-zero warning, 
# but the graph looks better if T0 is included, otherwise a joining line appears.
# (substitute 20 with 20.5 to avoid the warning in the graph)

Tgreaterthan20 = np.arange(20,60.5,0.05)
```

```python
T = np.concatenate((Tlessthan20, Tgreaterthan20), axis=None)
```

### Reference for an acceptable sensitivity: 0.25

```python
ref = np.full(len(T),0.25)
```

## Thermal exergy sensitivity graph ($\sigma$ vs T)
### Create a line in the graph for each deltaT0:

```python
for deltaT0 in deltaT0s:
    sigma = deltaT0/(T0 - T)
    plt.plot(T,sigma,'-',label=str(-deltaT0))
    
plt.xlabel('T ($^\circ$C)',fontsize=16,loc='right')
plt.ylabel('$\sigma$',fontsize=14,loc='top',rotation=0)
#plt.title('Thermal exergy sensitivity, T0 = 20$^\circ$C',fontsize=18)
    
plt.plot(T,ref,'--r')
plt.plot(T,-ref,'--b')
# Turn on the minor TICKS, which are required for the minor GRID
plt.minorticks_on()
plt.grid(True,which='both')
# Customize the major grid
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='grey')

plt.xlim(-10, 50)
plt.ylim(-0.5, 0.6)
plt.legend(loc=2,fontsize=10)

```
